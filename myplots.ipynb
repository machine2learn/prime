```python
import pandas as pd
import numpy as np
import matplotlib
# matplotlib.use('Agg')
%matplotlib inline
matplotlib.use('module://ipykernel.pylab.backend_inline')
import matplotlib.pyplot as plt
import math

import copy
import os
import shutil

%matplotlib inline
matplotlib.use('module://ipykernel.pylab.backend_inline')
import matplotlib.pyplot as plt
```

```python
n_samples = None
n_steps = 50
pairs_to_plot = [
    ['age_years.0', "z_global.2"],
     ["log_time_difference.0.2", "z_global.2"],
     ["time_difference.0.2", "z_global.2"],
     ["log_C_reactive_protein.0","z_global.2"],
     ["log_HbA1c_conv.0", "z_global.2"],
      ["log_MET_score.0", "z_global.2"],
      ["pain_tr.0", "z_global.2"],
     
    ["age_years.0", "z_change_snap_game_true_pos_rt_avrg.0.2"],
    ["log_time_difference.0.2", "z_change_snap_game_true_pos_rt_avrg.0.2"],
    ["time_difference.0.2", "z_change_snap_game_true_pos_rt_avrg.0.2"],
     ["log_C_reactive_protein.0","z_change_snap_game_true_pos_rt_avrg.0.2"],
      ["log_HbA1c_conv.0", "z_change_snap_game_true_pos_rt_avrg.0.2"],
      ["log_MET_score.0", "z_change_snap_game_true_pos_rt_avrg.0.2"],
       ["pain_tr.0", "z_change_snap_game_true_pos_rt_avrg.0.2"],
    
    ["age_years.0", "z_change_log_pairs_matching_sum_incorrect.0.2"],
    ["log_time_difference.0.2", "z_change_log_pairs_matching_sum_incorrect.0.2"],
     ["time_difference.0.2", "z_change_log_pairs_matching_sum_incorrect.0.2"],
       ["log_C_reactive_protein.0", "z_change_log_pairs_matching_sum_incorrect.0.2"],
       ["log_HbA1c_conv.0", "z_change_log_pairs_matching_sum_incorrect.0.2"],
        ["log_MET_score.0", "z_change_log_pairs_matching_sum_incorrect.0.2"],
        ["pain_tr.0", "z_change_log_pairs_matching_sum_incorrect.0.2"],
]
bin_colors = ["Diabetes_2.0"]
```

```python
input_file ="/projects/prime/ukbb/results_2024/pm_sg_0_2_no_interactions_quadr_hdl_ldl_remove_skewed_originals_13_05_HC1_gender_1/test_non_na.csv"
df = pd.read_table(input_file, sep=',', dtype=str, nrows=n_samples)

for clmn in df.columns:
    if clmn == "f.eid":
        continue
    if "date" in clmn:
        continue
    df[clmn] =  df[clmn].astype(float)
```

```python
for x_y_fld in pairs_to_plot:
    df_sorted = df.sort_values(x_y_fld[0]).reset_index()
    x = df_sorted[x_y_fld[0]]
    y = df_sorted[x_y_fld[1]]
    x_min = x.min()
    x_max = x.max()
    step = (x_max - x_min)/n_steps
    
    df_sorted['median'] = 0
    for i in range(len(df_sorted)):
            center = df_sorted.at[i, x_y_fld[0]]
            left = max(x_min, center -step)
            right = min(x_max,center+step )
            selected_rows = df_sorted[df_sorted[x_y_fld[0]].ge(left) & df_sorted[x_y_fld[0]].le(right)]
            v = selected_rows[x_y_fld[1]].median()
            df_sorted.at[i, 'median'] = v
    
    for bin_var in bin_colors:
        print(bin_var)
        colors = np.where(df_sorted[bin_var].astype(float) == 1,"red", "blue")
        plt.scatter(x, y, color = colors, s=5)
        plt.scatter(x, df_sorted['median'], color = "cyan", s=5)
        plt.xlabel(x_y_fld[0])
        plt.ylabel(x_y_fld[1])
        plt.legend(bin_var)
        plt.grid(False)
        plt.xticks([])
        plt.yticks([])

        # Save the plot as a PNG file
        # plt.savefig(f"{dir}/{predictor}.png", format='png', dpi=300)
        plt.show()
        
        
            
            
```

```python
df_test = copy.deepcopy(df).reset_index()
d_age = (df_test["age_years.0"].max() - df_test["age_years.0"].min())/n_steps
```

```python
d_hb1ac = (df_test["log_HbA1c_conv.0"].max() - df_test["log_HbA1c_conv.0"].min())/n_steps
d_td = (df_test["time_difference.0.2"].max() - df_test["time_difference.0.2"].min())/n_steps
```

```python
df_test['moving_median.z_global'] = None
for i in range(len(df_test)):
    center_age = df_test.at[i, 'age_years.0']
    center_hb1ac = df_test.at[i, 'log_HbA1c_conv.0']
    center_td = df_test.at[i, 'time_difference.0.2']
    selected_rows = df_test[df_test['age_years.0'].ge(center_age-d_age) & df_test['age_years.0'].ge(center_age+d_age)]
    selected_rows = selected_rows[selected_rows['log_HbA1c_conv.0'].ge(center_hb1ac-d_hb1ac) & selected_rows['log_HbA1c_conv.0'].ge(center_hb1ac+d_hb1ac)]
    selected_rows = selected_rows[selected_rows['time_difference.0.2'].ge(center_td-d_td) & selected_rows['time_difference.0.2'].ge(center_td+d_td)]
    df_test.at[i, 'moving_median.z_global'] = selected_rows["z_global.2"].median()

```

```python
for bin_var in bin_colors:
    print(bin_var)
    colors = np.where(df_test[bin_var] == 1,"orange", "cyan")
    
    plt.scatter(df_test['log_HbA1c_conv.0'], df_test['moving_median.z_global'], color = colors, s=5)
    plt.xlabel('log_HbA1c_conv.0')
    plt.ylabel('moving_median.z_global')
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
    plt.scatter(df_test['age_years.0'], df_test['moving_median.z_global'], color = colors, s=5)
    plt.xlabel('age_years.0')
    plt.ylabel('moving_median.z_global')
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.show()
    
    plt.scatter(df_test['time_difference.0.2'], df_test['moving_median.z_global'], color = colors, s=5)
    plt.xlabel('time_difference.0.2')
    plt.ylabel('moving_median.z_global')
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.show()
```
